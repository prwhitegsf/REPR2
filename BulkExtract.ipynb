```python

import src.FeatureExtractors as fe
import numpy as np
import pandas as pd

#cm = CreateRAVDESSMetadata()
df = pd.read_csv('datasets/RAVDESS/metadata/RAVDESS.csv',usecols=['actor','actor_sex','emotion','label','filepath'])
#df = df[df['filesize'] < 600000]
df.head()
```

```python
mfcc_list=[]

for row in df.itertuples():
    
    af = fe.AudioFeatures(row)
    mfcc_list.append(af.get_mfcc_as_npy(100))
   
mfcc_npy = np.array(mfcc_list)
with open('datasets/RAVDESS/features/mfcc/mfcc120.npy','wb') as f:
    np.save(f, mfcc_npy)
```

```python
num_filters = 16
outfile = f'datasets/RAVDESS/features/mel/mel{num_filters}.npy'

mel_list=[]

for row in df.itertuples():
    
    af = fe.AudioFeatures(row)
    mel_list.append(af.get_mel_as_numpy(num_filters))
   
mel_npy = np.array(mel_list)
with open(outfile,'wb') as f:
    np.save(f, mel_npy)
```

```python
feature_ds= np.load(outfile)
print(f'\nAudio samples represented: {feature_ds.shape[0]}')
print(f'Numerical features extracted per sample: {feature_ds.shape[1]}')
loaded_df = pd.DataFrame(feature_ds) # make it pretty for display
loaded_df
```

```python
df['features'] = list(feature_ds)
labels = np.array(df['label'])
features = np.vstack(df['features'])
print(f'\nAudio samples represented: {features.shape[0]}')
print(f'Numerical features extracted per sample: {features.shape[1]}')
ba_df = pd.DataFrame(features) # make it pretty for display
ba_df
```

```python
#features= mfcc_npy

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_scaled = features 
features_scaled = scaler.fit_transform(features_scaled)
```

```python
# Split Dataset
from sklearn.model_selection import train_test_split

############ Standard Scaled test/train set ###########
# The labels/classes (y_train, y_test) never change, keep old values 
feature_train, feature_test, label_train, label_test = train_test_split(
    features_scaled, 
    labels, 
    test_size=0.2, 
    random_state=69
)
```

```python
# Define the scoring function
from sklearn.model_selection import cross_validate
def print_model_scores(model, features, labels):
   
    scoring = ['recall_micro', 'precision_micro','accuracy']
    scores = cross_validate(model, features, labels,scoring=scoring)
    return pd.DataFrame(scores)
```

```python
# training and results
from sklearn.svm import SVC
from sklearn.metrics import recall_score, balanced_accuracy_score
# fowlkes_mallows_score
# positive_likelihood_ratio
model = SVC(
    gamma="scale",
    kernel='rbf',
    class_weight='balanced')

model.fit(feature_train, label_train)
#res = print_model_scores(model, feature_train, label_train)
#print_model_scores(model, X_train_scaled, y_train)
#print_model_scores(model,features_scaled, emotions)
p = model.predict(feature_test)

print(recall_score(label_test, p, average='micro'))
print(balanced_accuracy_score(label_test, p))
new_res = print_model_scores(model, feature_train, label_train)
#print(round(res['test_recall_micro'].mean(),3))
```
